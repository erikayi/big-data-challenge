```python
# import dependencies  
import pandas as pd
```

```python
# import csv files and save them as dataframe
jp_review = "Data/Review_Id_Data_AmazonReview_JP.csv"
jp_product = "Data/Products_Data_AmazonReview_JP.csv"
jp_total_customer = "Data/Total_Customers_By_ID_AmazonReview_JP.csv"
jp_vine = "Data/Vine_Table_AmazonReview_JP.csv"

jp_review_df = pd.read_csv(jp_review)
jp_product_df = pd.read_csv(jp_product)
jp_total_customer_df = pd.read_csv(jp_total_customer)
jp_vine_df = pd.read_csv(jp_vine)
```

```python
# join necessary dataframe to determine vine and other data analysis about the review
# joined review data and vine data 
joined_review_vine = jp_review_df.merge(jp_vine_df, how="inner")
joined_review_vine.head()
```

```python
# joined review data and product data 
joined_review_product = jp_review_df.merge(jp_product_df, how="inner")
joined_review_product.head()
```

```python
joined_review_customer = jp_review_df.merge(jp_total_customer_df, how="inner")
joined_review_customer.sort_values(by=["customers_count"], ascending=True)
```

```python
joined_review_vine = jp_review_df.merge(jp_vine_df, how="inner")
joined_review_vine.head()
```

```python
result = pd.concat([joined_review_product, joined_review_vine, joined_review_customer], 
                   axis=1, join="inner")
result
```
