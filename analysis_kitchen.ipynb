```python
# import dependencies  
import pandas as pd
```

```python
# import csv files and save them as dataframe
kitchen_review = "Data/Review_Id_Data_AmazonReview_Kitchen.csv"
kitchen_product = "Data/Products_Data_AmazonReview_Kitchen.csv"
kitchen_total_customer = "Data/Total_Customers_By_ID_AmazonReview_Kitchen.csv"
kitchen_vine = "Data/Vine_Table_AmazonReview_Kitchen.csv"

kitchen_review_df = pd.read_csv(kitchen_review)
kitchen_product_df = pd.read_csv(kitchen_product)
kitchen_total_customer_df = pd.read_csv(kitchen_total_customer)
kitchen_vine_df = pd.read_csv(kitchen_vine)
```
